Dataset: https://www.kaggle.com/datasets/vjchoudhary7/customer-segmentation-tutorial-in-python

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import pandas as pd
```

```python
df = pd.read_csv("/content/Mall_Customers.csv")
df.head()
```

```python
df = df[['Annual Income (k$)', 'Spending Score (1-100)']]
df.head()
```

```python
df.isna().sum()
```

```python
x = df.values
x
```

```python
c_list=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
  kmeans.fit(x)
  c_list.append(kmeans.inertia_)
plt.plot(range(1,11),c_list)
plt.title("The Elbow Method")
plt.xlabel("Number of cluster")
plt.ylabel("c_list")
plt.show()
```

```python
kmeans=KMeans(n_clusters=5,init='k-means++',random_state=42)
y_predict=kmeans.fit_predict(x)
y_predict
```

```python
plt.scatter(x[y_predict==0,0],x[y_predict==0,1],s=80,c="red",label="customer_1")
plt.scatter(x[y_predict==1,0],x[y_predict==1,1],s=80,c="blue",label="customer_2")
plt.scatter(x[y_predict==2,0],x[y_predict==2,1],s=80,c="yellow",label="customer_3")
plt.scatter(x[y_predict==3,0],x[y_predict==3,1],s=80,c="cyan",label="customer_4")
plt.scatter(x[y_predict==4,0],x[y_predict==4,1],s=80,c="black",label="customer_5")
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=100,c="magenta",label="Centroids")
plt.title("Clusters of customers")
plt.title("K-Means Clustering Algorithm")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
```
