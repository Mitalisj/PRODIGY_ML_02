```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import pandas as pd
```

```python
df = pd.read_csv("/content/Mall_Customers.csv")
df.head()
```

```python
df = df[['Annual Income (k$)', 'Spending Score (1-100)']]
df.head()
```

```python
df.isna().sum()
```

```python
x = df.values
x
```

```python
c_list=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
  kmeans.fit(x)
  c_list.append(kmeans.inertia_)
plt.plot(range(1,11),c_list)
plt.title("The Elbow Method")
plt.xlabel("Number of cluster")
plt.ylabel("c_list")
plt.show()
```

```python
kmeans=KMeans(n_clusters=2,init='k-means++',random_state=42)
y_predict=kmeans.fit_predict(x)
y_predict
```

```python
plt.scatter(x[y_predict==0,0],x[y_predict==0,1],s=100,c='blue',label='Cluster1')
plt.scatter(x[y_predict==1,0],x[y_predict==1,1],s=100,c='green',label='Cluster2')
plt.title("K-Means Clustering Algorithm")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
```
